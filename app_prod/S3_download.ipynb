```python
import boto3
import datetime

# Set up the S3 client
s3 = boto3.client('s3')

# Set the bucket name and storage class
bucket_name = config["cfg_s3_info"]["bucket"]
storage_class = 'BUCKET'  # or 'STANDARD_IA'

# Initialize the continuation token
next_token = None
all_files = [] 

# List the files in the bucket that are not stored in the Glacier storage class
while True:
    # Set the continuation token
    if next_token:
        kwargs = {'ContinuationToken': next_token}
    else:
        kwargs = {}

    # Retrieve the next page of results
    response = s3.list_objects_v2(
        Bucket=bucket_name,
        Prefix = "ROC/PROCESSED_DATA",
        **kwargs
    )

    # Print the file names
    for file in response['Contents']:
        if file["StorageClass"] !="GLACIER":
            try:
                if "LAST" in file["Key"]:
                    continue
                all_files.append(file["Key"])
            except Exception as e:
                print(f"Error for file: {file['Key']}")
                print(e)
            


    # Update the continuation token
    next_token = response.get('NextContinuationToken')

    # Stop when there are no more pages
    if not next_token:
        break
```
